```python
import pandas as pd
import numpy as np
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import nltk 
import string 
import re
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem import WordNetLemmatizer 
```

```python
def preprocess_text(messages):
    #lowercase
    messages = messages.lower()
    #punctuation
    translator = str.maketrans('', '', string.punctuation)
    messages = messages.translate(translator)
    #stopwords
    stop_words = set(stopwords.words("english")) 
    word_tokens = word_tokenize(messages)
    filtered_messages = [word for word in word_tokens if word not in stop_words]
    #lemmatize
    lemmatizer = WordNetLemmatizer() 
    f_messages = [lemmatizer.lemmatize(word, pos ='v') for word in filtered_messages]

    return " ".join(f_messages)
```

```python
df_merged = df_merged.dropna()
```

```python
df_merged
```

```python
df_merged['preprocess_lyrics'] = df_merged['lyrics'].apply(preprocess_text)
```

```python
df_merged
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
desc_blob = [TextBlob(lyrics) for lyrics in df_merged['preprocess_lyrics']]
#add the sentiment metrics to the dataframe
df_merged['tb_Pol'] = [b.sentiment.polarity for b in desc_blob]
df_merged['tb_Subj'] = [b.sentiment.subjectivity for b in desc_blob]
```

```python
df_merged
```

```python
#load VADER
analyzer = SentimentIntensityAnalyzer()
#Add VADER metrics to dataframe
df_merged['compound'] = [analyzer.polarity_scores(v)['compound'] for v in df_merged['preprocess_lyrics']]
df_merged['neg'] = [analyzer.polarity_scores(v)['neg'] for v in df_merged['preprocess_lyrics']]
df_merged['neu'] = [analyzer.polarity_scores(v)['neu'] for v in df_merged['preprocess_lyrics']]
df_merged['pos'] = [analyzer.polarity_scores(v)['pos'] for v in df_merged['preprocess_lyrics']]
```

```python
df_merged
```

```python
df_merged = df_merged.drop(columns=["artist","title"])
```

```python
df_final = df3.merge(df_merged, how='outer', left_index=True, right_index=True)
```
